```python
import random 

########################################### Sender Program ###############################################
class sender:
        
        #Program to make frame of random size.
        
    def frames(self,low,high):
        f_input=open("input.txt","r")
        data=f_input.read()
        string=list(data)
        f_input.close()
        
        print(f"\n\nOriginal File data")
        print("_________________________________________________________________________\n")
        print(data)
        
        count=0
        f_frame=open("frame.txt","w")
        f_frame.write("Framing the input According to Size Given by user!\n")  
        f_frame.close()
        
        f_frame=open("frame.txt","a")
        print("\n\n#########################################################################")
        print("Framing the input according to random limit given by user!")
        print("#########################################################################")
        while( (len(string)-count) > high - low ):
            size=random.randint(low,high)
         
            frame=''
            frame=''.join(string[count:size+count])
            print(frame)
            f_frame.write(frame+"\n")
            count+=size
            
        frame=''
        frame=''.join(string[count:size+count])
        print(frame)
        print("#########################################################################\n")
        f_frame.write(frame+"\n")
        
        f_frame.close()
        
#code for byte stuffing by introducing flag and escape character at appropriate places.
    def byte_stuffing(self,flag,esc):
        
        f_input=open("frame.txt","r")
        f_input.readline()
        string=f_input.read()
        f_input.close()
        
        f_byte=open("packet.txt","w")
        f_byte.write("Frames Along with Byte Stuffing\n")
        f_byte.close()
        
        stream=''
        
        print("\n*************************************************************************")
        print(f"Frames Along with Byte Stuffing!\t Flag={flag}\t Escape={esc}")
        print("*************************************************************************")
        
        for frame in string.split('\n'):
            if (frame != ''):
                temp=''

                for i in range(len(frame)):
                    if(frame[i]==flag or frame[i]==esc):
                        temp=temp + esc + frame[i]
                    else:
                        temp=temp+frame[i]
                
                f_byte=open("packet.txt","a")
                f_byte.write(flag+temp+flag)
                
                stream+=flag+temp+flag
                
                print(flag+temp+flag)
        print("*************************************************************************\n\n")
        
        print("Sender Sends the message...")
        print("_________________________________________________________________________")
        print(stream)
        print()
        print()
        
        f_byte.close()
        return
    
#Program to convert Byte stream to bit stream.
    def bit_conversion(self):
        f_input=open("input.txt","r")
        data=f_input.read()
        f_input.close()
        
        print(f"\n\nOriginal File data")
        print("_________________________________________________________________________\n")
        print(data)
        
        binary = " ".join(f"{ord(i):07b}" for i in data)
        ls=(binary.split())
        binary=''.join(ls)
        
        f_binary=open("binary.txt","w")
        f_binary.write(binary)
        f_binary.close()
        
        print(f"\nBinary Converted data")
        print("_________________________________________________________________________\n")
        print(binary)
        return
        
#Program to do bit stuffing by adding one zero after every 5 consecutive 1's.
    
    def bit_stuffing(self):
        f_binary=open("binary.txt","r")
        data=list(f_binary.read())
        
        i=0
        c=0
        
        while (i<len(data)):
            if(data[i]=='1'):
                c+=1

            else: c=0

            if(c==5):
                data.insert(i+1,'0')
                c=0
            i+=1

        x=''
        x=''.join(data)
        
        f_packet=open("packet.txt","w")
        f_packet.write(x)
        f_packet.close()
        
        print("\nBit stuffed data (data to be send)")
        print("_________________________________________________________________________\n")
        print(x)
        print("_________________________________________________________________________\n")
        
        print("Sender Sends the message...\n\n")
        return  
        
```
